```python
import cv2
import easyocr
import matplotlib.pyplot as plt
import numpy as np

# read image
image_path = '/asa_passport.jpeg'
img = cv2.imread(image_path)

# instance text detector
reader = easyocr.Reader(['en'], gpu=False)

# detect text on image
text_ = reader.readtext(img)

# Sort detected text based on the position (y-coordinate) of the bounding box
text_sorted = sorted(text_, key=lambda x: x[0][0][1])

threshold = 0.25
prev_y = None
line_text = ""

# print the detected text in the order they appear and draw bbox and text
for t in text_sorted:
    bbox, text, score = t

    if score > threshold:
        pt1 = (int(bbox[0][0]), int(bbox[0][1]))
        pt2 = (int(bbox[2][0]), int(bbox[2][1]))

        # New line detection based on vertical spacing (y-coordinate change)
        if prev_y is not None and abs(prev_y - pt1[1]) > 20:  # Adjust this threshold based on image characteristics
            print(line_text.strip())
            line_text = ""  # Reset line text for the new line

        line_text += text + " "
        prev_y = pt1[1]

        # Draw bounding box and text on the image
        cv2.rectangle(img, pt1, pt2, (0, 255, 0), 5)
        cv2.putText(img, text, pt1, cv2.FONT_HERSHEY_COMPLEX, 0.65, (255, 0, 0), 2)

# Print the last accumulated line
if line_text:
    print(line_text.strip())

# Show image with bounding boxes
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Hide axes for better view
plt.show()

```

```python
# for local storage + OCR


import os
import cv2
import easyocr
import matplotlib.pyplot as plt
import numpy as np
from tkinter import Tk, filedialog

def select_directory():
    """
    Opens a dialog for the user to select a directory.
    Returns the selected directory path.
    """
    root = Tk()
    root.withdraw()
    directory = filedialog.askdirectory(title="Select Directory to Scan")
    root.destroy()
    return directory

def scan_local_storage(directory):
    """
    Scans the specified directory and returns a list of image file paths.
    Only includes files with common image extensions.
    """
    image_extensions = {'.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.tif'}
    file_paths = [
        os.path.join(root, file)
        for root, _, files in os.walk(directory)
        for file in files
        if os.path.splitext(file)[1].lower() in image_extensions
    ]
    return file_paths

def analyze_image(image_path):
    """
    Analyzes an image file using OCR and displays the image with detected text.
    """
    # Load the image
    img = cv2.imread(image_path)

    # Initialize EasyOCR reader
    reader = easyocr.Reader(['en'], gpu=False)

    # Detect text on the image
    text_ = reader.readtext(img)

    # Sort detected text based on the y-coordinate of the bounding box
    text_sorted = sorted(text_, key=lambda x: x[0][0][1])

    threshold = 0.25
    prev_y = None
    line_text = ""

    # Print and draw bounding boxes around detected text
    for t in text_sorted:
        bbox, text, score = t
        if score > threshold:
            pt1 = (int(bbox[0][0]), int(bbox[0][1]))
            pt2 = (int(bbox[2][0]), int(bbox[2][1]))

            # New line detection based on y-coordinate change
            if prev_y is not None and abs(prev_y - pt1[1]) > 20:
                print(line_text.strip())
                line_text = ""

            line_text += text + " "
            prev_y = pt1[1]

            # Draw bounding box and text on the image
            cv2.rectangle(img, pt1, pt2, (0, 255, 0), 5)
            cv2.putText(img, text, pt1, cv2.FONT_HERSHEY_COMPLEX, 0.65, (255, 0, 0), 2)

    # Print the last accumulated line
    if line_text:
        print(line_text.strip())

    # Display the image with bounding boxes
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

def main():
    """
    Main function to prompt directory selection, scan for images, and analyze them.
    """
    # Step 1: Ask the user to select a directory
    directory = select_directory()
    if not directory:
        print("No directory selected. Exiting...")
        return

    # Step 2: Get a list of image files in the directory
    image_files = scan_local_storage(directory)

    if not image_files:
        print("No image files found in the selected directory.")
        return

    # Step 3: Analyze each image file
    for image_path in image_files:
        print(f"Analyzing {image_path}...")
        analyze_image(image_path)

# Run the main function
main()
```
